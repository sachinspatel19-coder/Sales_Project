```python
df3.drop(columns=['Notes','Status'],inplace=True)
```

```python
df3.columns
```

```python
df3['Benefits']
df3['Benefits']=df3['Benefits'].bfill()
```

```python
df3['Benefits'].isnull().sum()
```

```python
df3['BasePay']=df3['BasePay'].fillna(df3['BasePay'].mean)
```

```python
df3['OvertimePay']=df3['OvertimePay'].fillna(df3['OvertimePay'].median)
```

```python
df3['OtherPay']=df3['OtherPay'].fillna(df3['OtherPay'].mode)
```
